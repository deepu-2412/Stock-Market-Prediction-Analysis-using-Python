Task 5: calculating moving average

```python
df
```

```python
df1=df['Close'].to_frame()
df1['SMA30C']=df['Close'].rolling(30).mean()
df1.dropna(inplace=True)
df1
```

```python
df1[['Close', 'SMA30C']].plot(label='AMAZON',figsize=(16, 8))
plt.title('MOVING AVERAGE OF CLOSE STOCK')
plt.show()
```

```python
#MOVING AVERAGE FOR OPEN STOCK
df2=df['Open'].to_frame()
df2['SMA30O']=df['Open'].rolling(30).mean()
df2.dropna(inplace=True)
df2
```

```python
df2[['Open', 'SMA30O']].plot(label='AMAZON',figsize=(16, 8))
plt.title('MOVING AVERAGE OF OPEN STOCK')
plt.show()
```

TASK 7:Add a new column ‘Trend’ whose values are based on the 'Daily Return'.

```python
df['Trend']=amazon_daily_returns
df
```

calculating percentage gain or loss


(closing-opening)/opening x100 to find percentage of gain or loss each day


```python
df['per_tre']=(df.Trend/df.Open)*100
df
```

TASK 8:Visualize trend frequency through a Pie Chart.

```python
def freq(x):
    if x['per_tre']>2:
        return 'gain >2%'
    elif 0<x['per_tre']<2:
        return 'gain 0-2%'
    elif 0>x['per_tre']>-2:
        return 'loss 0-2%'
    else:
        return 'loss >2%'
```

```python
df['tre_fre']=df.apply(freq,axis=1)
```

```python
df
```

```python

df["dummy"]=1

df.groupby(['tre_fre']).sum().plot(kind='pie',y='dummy',title='total gain/loss frequency(Trend Frequency)',autopct='%1.0f%%')
```
